# Generate undersampled data as train and test data

```python


def getData(trnTst = 'testing', num = 100, sigma = 0.01):
    '''
    The value of num could be set between 0 and 163. The test pieces have 164 in total.
    
    '''
    filename = 'dataset.hdf5'
    with h5.File(filename) as f:
        if trnTst=='training':
            org,csm,mask=f['trnOrg'][:],f['trnCsm'][:],f['trnMask'][:]
        else:
            org,csm,mask=f['tstOrg'][num],f['tstCsm'][num],f['tstMask'][num]
            na=np.newaxis
            org,csm,mask=org[na],csm[na],mask[na]
    print('Successfully read the data from file.')
    print('Now doing undersampling...')
    data_undersampled = generateUndersampled(org, csm, mask, sigma)
    print('Successfully undersampled data!')


    # For training = True:
    # org: nSlices * nrow * ncol
    # data_undersampled : nSlices * nrow * ncol
    # csm: nSlices * ncoil * nrow * ncol
    # mask: nSlices * nrow * ncol
    return org,data_undersampled,csm,mask

def A_oper(Img, csm, mask, nrow, ncol, ncoil):
    """
    This is the A operator as defined in the paper.

    Function description:

    Img : nrow * ncol

    csm: ncoil * nrow * ncol

    mask: nrow * ncol

    """

    # After np.tile operation, Img would have a shape of [ncoil, nrow, ncol], and the csm here should have the shape of [ncoil, nrow, ncol]
    coil_Images = np.tile(Img, [ncoil, 1, 1]) * csm

    kspace = np.fft.fft2(coil_Images) / np.sqrt(nrow * ncol)
    # kspace would have the same shape of [ncoil, nrow, ncol] and the spatial domain is converted to frequency domain.

    if len(mask.shape) == 2:
        mask = np.tile(mask, [ncoil, 1, 1])

    # masking out the values according to the mask in the frequency domain (Flattened!)
    res = kspace[mask != 0]

    ########Note: Here the spatial shape changes############
    return res

def transpose_A_oper(k_spaceUnder, csm, mask, nrow, ncol, ncoil):
    """
    Function description:

    k_spaceUnder: Undersampled k data which is the output of A_oper

    csm: ncoil * nrow * ncol

    mask: nrow * ncol
    """
    temp = np.zeros((ncoil, nrow, ncol), dtype = np.complex64)

    if len(mask.shape) == 2:
        mask = np.tile(mask, (ncoil, 1, 1))
    
    temp[mask!=0] = k_spaceUnder

    # shape of (ncoil * nrow * ncol)
    img = np.fft.ifft2(temp) * np.sqrt(nrow * ncol)

    coilComb = np.sum(img * np.conj(csm), axis = 0).astype(np.complex64)

    # coilComb has the shape of nrow * ncol
    return coilComb

def generateUndersampled(org, csm, mask, sigma=0.):

    """
    Function description:

    org: nSlice * nrow * ncol

    csm: nslice * ncoil * nrow * ncolumn

    mask: nslice * nrow * ncolumn
    """

    nSlice, ncoil, nrow, ncol= csm.shape

    atb=np.empty(org.shape, dtype=np.complex64)

    for i in range(nSlice):
        A  = lambda z: A_oper(z, csm[i], mask[i], nrow, ncol, ncoil)
        At = lambda z: transpose_A_oper(z, csm[i], mask[i], nrow, ncol, ncoil)
        
        # returns the array of the indices of the flattened array where the value is Non-zero
        sidx=np.where(mask[i].ravel()!=0)[0]
        nSIDX=len(sidx)

        noise=np.random.randn(nSIDX*ncoil,)+1j*np.random.randn(nSIDX*ncoil,)

        noise=noise*(sigma/np.sqrt(2.))
        
        y=A(org[i]) + noise
        # Add noise to the unmasked k-space data

        atb[i]=At(y)
        
        # atb has shape of nSlice * nrow * ncol
    return atb
```

# This is a function that could be tested on the train and test to test the restoration performance

```python
def myPSNR(org, recon):
    """ This function calculates PSNR between the original and
    the reconstructed images
    
    Function description:

    org: nrow * ncol

    recon: nrow * ncol
    
    """

    #org.size(): nrow * ncol
    mse=np.sum(np.square( np.abs(org-recon)))/org.size
    psnr=20*np.log10(org.max()/(np.sqrt(mse)+1e-10 ))
    return psnr
```

# Create the CNN layer

```python
class createLayer(keras.Model):
    """
    This block creates a standard convolution module 
    
    """
    def __init__(self, num_of_filters, filter_shape, strides = (1,1), padding = 'same',  lastLayer = False, **kwargs):
        super().__init__(**kwargs)
        self.num_of_filters = num_of_filters
        self.filter_shape = filter_shape
        self.lastLayer = lastLayer
        self.strides = strides
        self.padding = padding


        self.Conv2D = keras.layers.Conv2D(filters = self.num_of_filters,
                kernel_size = self.filter_shape,
                strides = self.strides,
                padding = self.padding,
                kernel_initializer='glorot_uniform')

        self.BN = keras.layers.BatchNormalization()

    def call(self, input):

        res = self.Conv2D(input)
        res = self.BN(res)
        
        if not (self.lastLayer):

            return keras.layers.ReLU()(res)
            
        else:

            return res
        
```

```python
class Denoiser(keras.Model):
    """

    This is the denoiser block that uses Denoiser block just designed

    """
    def __init__(self, block = createLayer, **kwargs):

        """
        block is at default set to createLayer

        """
        super().__init__(**kwargs)
        self.block = block

        # nLay is the number of layers that consist this standard Denoiser block
        # In this implementation, we could set it by explicitly adjusting it here.
        self.layer1 = self.block(num_of_filters = 64, filter_shape = (3,3), lastLayer = False)
        self.layer2 = self.block(num_of_filters = 2, filter_shape = (3,3), lastLayer = True)
        
    def call(self, input):

        """
        input : [batch_size, nrow, ncol, 2]

        """

        short_cut = tf.identity(input)

        res = short_cut
    
        res = self.layer1(res)

        res = self.layer2(res)


        # shape of [batch_size, nrow, ncol, 2]  
        return tf.add(short_cut, res)
            
```

# This is the data consistency block

```python
class Data_consistency_CG(keras.layers.Layer):
    def __init__(self, **kwargs):

        """
        SF : Normalization factor
        """
        super().__init__(**kwargs)

        self.lam1 = tf.Variable(name = 'lam1', dtype = tf.float32, initial_value = tf.constant(0.05), trainable = True, shape = None)
        
    def real2complex(self, x):
        
        return tf.cast(tf.complex(x[..., 0], x[..., 1]), tf.complex64)

    def complex2real(self, x):

        return tf.cast(tf.stack([tf.math.real(x), tf.math.imag(x)], axis = -1), tf.float32)

    def transform_A(self, img, csm, mask):
        
        # img : [nrow, ncol] with dtype : complex64
        # csm : [ncoil, nrow, ncol]
        # mask : [nrow, ncol]

        nrow = tf.shape(csm)[1]
        ncol = tf.shape(csm)[2]

        pixels = nrow * ncol

        SF = tf.complex(tf.sqrt(tf.cast(pixels, tf.float32)), 0.)


        coilImages = csm * img
        # shape of (ncoil, nrow, ncol)


        kspace = tf.signal.fft2d(input = coilImages) / SF
        # shape of (ncoil, nrow, ncol)
    
        temp=kspace* mask
        # shape of (ncoil, nrow, ncol)

        coilImgs =tf.signal.ifft2d(temp)* SF
        # shape of (ncoil, nrow, ncol)

        coilComb= tf.reduce_sum(coilImgs*tf.math.conj(csm), axis= 0)
        # shape of (nrow, ncol)

        coilComb=coilComb + tf.cast(self.lam1, tf.complex64) * img
        # shape of (nrow, ncol)

        return coilComb
    
    
    def CG(self, input):
        """
        termi_cond : the termination condition for the while loop

        i : the iteration number
        r : residual value
        rTr : the square absolute value of r
        p : the p is direcion inside the CG algorithm
        
        img: [nrow * ncol * 2]
        csm: [ncoils * nrow * ncol]
        mask: [nrow * ncol]
        """


        img = input[0]
        csm = input[1]
        mask = input[2]

        csm = tf.cast(csm, tf.complex64)
        mask = tf.cast(mask, tf.complex64)

        rhs = self.real2complex(img)
        # shape of (nrow, ncol)

        termi_cond = lambda i, rTr, *_ : tf.logical_and(tf.less(i, 10), rTr > 1e-10)
        
        
        def CG_body(i, rTr, x, r, p):
            '''
            i : scalar
            rTr : scalar
            r : nrow * ncol (complex)
            x : nrow * ncol (complex)
            p : nrow * ncol (complex)
            '''
            # Note: p is set to r at the initialization part later.
            Ap = self.transform_A(p, csm, mask)
            alpha = rTr / tf.cast(tf.reduce_sum( tf.math.conj(p) * Ap ), tf.float32)
            alpha = tf.complex( alpha, 0. )
            x = x + alpha * p
            r = r - alpha * Ap
            rTrNew = tf.cast( tf.reduce_sum(tf.math.conj(r) * r), tf.float32)
            beta = rTrNew / rTr
            beta = tf.cast(beta, tf.complex64)
            p = r + beta * p
            return i+1, rTrNew, x, r, p

        x = tf.zeros_like(rhs)
        #initialize i(index), r(residual), p(direction)
        i, r, p = 0, rhs, rhs
        rTr = tf.cast(  tf.reduce_sum(tf.math.conj(r) * r), tf.float32)

        loopVar = (i, rTr, x, r, p)

        out = tf.while_loop(termi_cond, CG_body, loopVar, parallel_iterations = 1)

        return self.complex2real(out[2])
        # output shape: [nrow, ncol, 2]

    def apply_CG_each_image_batch(self, input):
        '''
        This function applies CG on each image in the batch
        
        '''
        
        batch_image_holder = tf.map_fn(self.CG, input, fn_output_signature = tf.float32)

        return batch_image_holder


    def call(self, input):
        """

        inputs: Parameters of dimension 
        
        img : [batch_size, nrow, ncol, 2] * csm : [batch_size, ncoils, nrow , ncol] * mask : [batch_size, nrow, ncol]

        """



        y = self.apply_CG_each_image_batch(input)
        # output shape: [batch_size, nrow, ncol, 2]
        
        return y

            
            
```

```python
class Model(keras.Model):
    def __init__(self,
                denoiser = Denoiser, 
                dc_block = Data_consistency_CG,
                **kwargs):


        super().__init__(**kwargs)

        
        self.lam2 = tf.Variable(name = 'lam2', dtype = tf.float32, initial_value = tf.constant(0.05), trainable = True, shape = None)

        self.denoiser = denoiser(block = createLayer)
        self.dc_block = dc_block()


    def call(self, inputs, training):
        """
        Inputs: Undersampled data with Gaussian noise added
                csm : [batch_size, ncoils, nrow , ncol] * mask : [batch_size, nrow, ncol] * img : [batch_size, nrow, ncol, 2]

        """

        batch_csm = inputs[0]
        batch_mask = inputs[1]
        batch_img = inputs[2]

        # res = tf.identity(batch_img, name = 'inputs')
            
        res = self.denoiser(batch_img)
        # return the noise patterns with a shape of (batch_size, nrow, ncol, 2)

        res = self.lam2 * res + batch_img
        # This returns the rhs values (The denoised values) with a shape of (batch_size, nrow, ncol, 2)

        res = self.dc_block([res, batch_csm, batch_mask])

        #res has a dimension of (batch_size, nrow, ncol, 2) at this step

        ###########################################
        #       Only the most basic block         #
        #     Only for testing purposes           #
        #      Ignore what is written below       #
        ###########################################

        # res = self.denoiser(res)
        # # returns the output of the denoised images with dimension of (batch_size, nrow, ncol, 2)

        # res = self.lam1 * res + inputs
        #     # This returns the rhs values (The denoised values) with a shape of (batch_size, nrow, ncol, 2)

        # res = self.dc_block(res)

        #res has a dimension of (batch_size, nrow, ncol, 2) at this step


        return res

```

```python
def r2c(inp):
    """  
    input img: row x col x 2 in float32
    output image: row  x col in complex64

    """
    if inp.dtype=='float32':
        dtype=np.complex64
    else:
        dtype=np.complex128
    out=np.zeros( inp.shape[0:2],dtype=dtype)
    out=inp[...,0]+1j*inp[...,1]
    return out

def c2r(inp):
    """  
    input img: n_slices * row * col in complex64
    output image: n_slices * row  * col * 2 in float32
    
    """
    
    if inp.dtype=='complex64':
        dtype=np.float32
    else:
        dtype=np.float64
    
    real = np.real(inp).astype(dtype)
    imag = np.imag(inp).astype(dtype)

    out = np.stack((real, imag), axis = -1)

    return out

def real2complex(self, x):
        
    return tf.complex(x[..., 0], x[..., 1])

def complex2real(self, x):

    return tf.stack([tf.real(x), tf.imag(x)], axis = -1)
```
